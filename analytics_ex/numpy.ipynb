# 2차원 데이터 만들기

```python
a2d = [[1,2,3], [4,5,6], [7,8,9]]
a2d
```

```python
type(a2d)
```

```python
a2d = np.array(a2d)
a2d
```

```python
# 2차원 -> 행의 갯수
len(a2d)
```

```python
# array의 전체 요소의 갯수
a2d.size
```

```python
a2d.shape
```

# Save and Load data

```python
a2d
```

```python
# 파일로 저장하기
np.save('my_array', a2d)
```

```python
# 파일로 저장된것 불로오기
arr2d = np.load('my_array.npy')
arr2d
```
