# seed값을 5로 고정한 후 random.randint(1,11)을 세번 실행했을 때 나오는 값을 구하시오

```python
np.random.seed(5)
np.random.randint(1,11)
np.random.randint(1,11)
np.random.randint(1,11)
```

```python
import random
random.seed(5)
print(random.randint(1, 11))
print(random.randint(1, 11))
print(random.randint(1, 11))
```
